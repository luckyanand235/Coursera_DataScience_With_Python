```python
import pandas as pd
```

```python
df = pd.read_csv('C:\DataAnalytics\olympics.csv', index_col=0, skiprows=1)
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns = {col:'Gold'+col[4:]}, inplace=True)
    if col[:2] == '02':
        df.rename(columns = {col:'Silver'+col[4:]}, inplace=True)
    if col[:2] == '03':
        df.rename(columns = {col:'Bronze'+col[4:]}, inplace=True)
    if col[:1] == '№':
        df.rename(columns = {col:'#'+col[1:]}, inplace=True)
names_ids = df.index.str.split('\s\(') # split the index by '('
```

```python
import numpy as np
```

```python
df.index = names_ids.str[0]
df['ID'] = names_ids.str[1].str[:3]
```

```python
df = df.drop('Totals')
```

```python
df.head()
```

```python
def first_country():
    return df.iloc[0]
first_country()  # first country of the data frame
```

```python
def max_gold():
    return df['Gold'].idxmax()  # max gold medals in summer olympics
max_gold()
```

```python
def diff():
    df['Diff'] = abs(df['Gold'] - df['Gold.1']) # creating a difference series in data frame in order to find the absolute difference between the summer and the winter olymipcs of the counturie
    return df['Diff'].idxmax()  # and getting the maximum value out of it
diff()
```

```python
def ans_3():
    df_res = df.copy()
    df_res = df_res[(df_res['Gold.1'] > 0) & (df_res['Gold']>0) ]
    df_res['Diff'] = abs(df_res['Gold'] - df_res['Gold.1'])
    df_res['Total Gold'] = df_res['Gold'] + df_res['Gold.1'] + df_res['Gold.2']
    df_res['Relative Gold'] = df_res['Diff']/df_res['Total Gold']
    return df_res['Relative Gold'].idxmax()
ans_3()
```

```python
def ans_4():
    points = df['Gold.2']*3 + df['Silver.2']*2 + df['Bronze.2']
    return points
ans_4()
```

```python
census_df = pd.read_csv('C:\DataAnalytics\census.csv')
```

```python
census_df.head()
```

```python
def ans_5():
    res = census_df.copy()
    res = res.reset_index()
    res = res[res['SUMLEV'] == 50]
    res = res.set_index(['STNAME', 'CTYNAME'])
    
    res = res.reset_index(level=1).index.value_counts().idxmax()
    return res

ans_5()
```

```python
def ans_6():
    res = census_df.copy()
    res = res.reset_index()
    res = res[res['SUMLEV'] == 50]
    cols_keep = ['STNAME', 'CENSUS2010POP']
    res = res[cols_keep]
    res = res.sort_values(['STNAME', 'CENSUS2010POP'], ascending=False)
    res = res.groupby('STNAME').head(3)
    res = res.groupby("STNAME").sum().sort_values('CENSUS2010POP', ascending=False).reset_index()
    res = list(res['STNAME'].loc[:2])
    return res
ans_6()
```

```python
def ans_7():
    res = census_df.copy()
    res = res.reset_index()
    res = res[res['SUMLEV'] == 50]
    using_cols = ['POPESTIMATE2010',
                 'POPESTIMATE2011',
                 'POPESTIMATE2012', 
                 'POPESTIMATE2013',
                 'POPESTIMATE2014',
                 'POPESTIMATE2015']
    res['MinPop'] = res.loc[:, using_cols].min(axis=1)
    res['MaxPop'] = res.loc[:, using_cols].max(axis=1)
    res['PopDelta'] = res['MaxPop'] - res['MinPop']
    keep_cols = ['CTYNAME', 'PopDelta']
    res = res[keep_cols].sort_values('PopDelta', ascending=False).reset_index()
    res = res['CTYNAME'].loc[0]
    return res

ans_7()
```

```python
def ans_8():
    res = census_df.copy()
    res = res.reset_index()
    res = res[res['SUMLEV'] == 50]
    res = res[(res['REGION'] <= 2)]
    res = res[(res['POPESTIMATE2015'] > res['POPESTIMATE2014'])]
    res = res[(res['CTYNAME'].str[0:10] == 'Washington')].sort_index()
    keep_cols = ['STNAME', 'CTYNAME']
    res = res[keep_cols]
    
    return res

ans_8()
```
